```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from matplotlib.pyplot import figure
f1=open("Records5avl.txt",'r')
f2=open("Records5veb.txt",'r')
lines1=f1.readlines()
lines2=f2.readlines()
f1.close()
f2.close()
x1=[]
lst1=[]
lst2=[]
for line1 in lines1:
    x1.append(int(line1.split()[0]))
    lst1.append(int(line1.split()[1]))
for line2 in lines2:
    lst2.append(int(line2.split()[1]))
df1=pd.DataFrame({'x11':x1,'y1':lst1})
df2=pd.DataFrame({'x12':x1,'y2':lst2})
#figure()
fig=plt.figure(num=None, figsize=(10,12), dpi=100, facecolor='w', edgecolor='k')
plt.plot('x11','y1',data=df1,color='violet')
plt.plot('x12','y2',data=df2,color='red')
plt.xlabel('Number of Edges-->',fontsize=16)
plt.ylabel('Execution time in microseconds',fontsize=16)
fig.savefig('avlveb.jpg')
plt.show()
```
